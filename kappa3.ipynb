```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import cohen_kappa_score
from sklearn.preprocessing import StandardScaler

# Load dataset with error handling
def load_dataset(filepath):
    try:
        # Try UTF-8 encoding first
        df = pd.read_csv(filepath, sep='\t', encoding='utf-8')
    except UnicodeDecodeError:
        # Fall back to latin1 if UTF-8 fails
        df = pd.read_csv(filepath, sep='\t', encoding='latin1')
    except FileNotFoundError:
        print(f"File not found: {filepath}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
    return df

# Load dataset
df = load_dataset('dataset.tsv')

# Check if DataFrame is loaded successfully
if df is None:
    raise ValueError("Failed to load dataset. Check the file path and encoding.")

# Select features and target
features = ['rater1_trait1', 'rater1_trait2', 'rater1_trait3', 'rater1_trait4', 'rater1_trait5', 'rater1_trait6',
             'rater2_trait1', 'rater2_trait2', 'rater2_trait3', 'rater2_trait4', 'rater2_trait5', 'rater2_trait6',
             'rater3_trait1', 'rater3_trait2', 'rater3_trait3', 'rater3_trait4', 'rater3_trait5', 'rater3_trait6']
target = 'domain1_score'

# Ensure features and target are in DataFrame
if not all(col in df.columns for col in features) or target not in df.columns:
    raise ValueError("One or more feature/target columns are missing in the DataFrame.")

X = df[features].copy()
y = df[target].copy()

# Handle missing values
X.fillna(X.mean(), inplace=True)
y.fillna(y.mean(), inplace=True)

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize models
models = {
    'SVR': SVR(),
    'KNN': KNeighborsRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Neural Network': MLPRegressor(max_iter=500)
}

kappa_scores = {}

# Train and evaluate models
for name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    kappa_score = cohen_kappa_score(y_test, np.round(predictions))
    kappa_scores[name] = kappa_score

    # Plot each model's predictions vs. actual values
    plt.figure(figsize=(10, 6))
    plt.scatter(y_test, predictions, alpha=0.5)
    plt.title(f'{name} Predictions vs Actual Values')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.grid(True)
    plt.show()

# Plot bar chart of kappa scores
plt.figure(figsize=(10, 6))
plt.bar(kappa_scores.keys(), kappa_scores.values(), color='skyblue')
plt.title('Model Kappa Score Comparison')
plt.xlabel('Model')
plt.ylabel('Kappa Score')
plt.ylim(0, 1)  # Assuming kappa scores are between 0 and 1
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
